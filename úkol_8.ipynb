1. Vypi≈° si prvn√≠ch nƒõkolik ≈ô√°dk≈Ø, a≈• si prohl√©dne≈° strukturu tabulky.

```python
import pandas as pd
seznam = pd.read_csv("temperature.csv", sep=",")
seznam.head()
```

2. Dotaz na mƒõ≈ôen√≠, kter√° byla provedena v Praze. Je na datech nƒõco zvl√°≈°tn√≠ho?

```python
praha = seznam[seznam["City"] == "Prague"]
print("\n Hodnoty namƒõ≈ôen√© v Praze jsou: ")
praha.head()
```

3. Dotaz na mƒõ≈ôen√≠, ve kter√Ωch je teplota (sloupec AvgTemperature) vy≈°≈°√≠ ne≈æ 80 stup≈à≈Ø.

```python
teplota_nad_80 = seznam[seznam["AvgTemperature"] > 80]
print("\n Mƒõ≈ôen√≠ u kter√Ωch byla teplota vy≈°≈°√≠ nƒõ≈æ 80: ")
teplota_nad_80.head()
```

4. Dotaz na mƒõ≈ôen√≠, ve kter√Ωch je teplota vy≈°≈°√≠ ne≈æ 60 stup≈à≈Ø a souƒçasnƒõ bylo mƒõ≈ôen√≠ provedeno v regionu (sloupec Region) Evropa (Europe).

```python
vysok√°_teplota_evropske_staty = seznam[
    (seznam["AvgTemperature"] > 60) & (seznam["Region"] == "Europe")
]
print("\n Evropsk√© st√°ty u kter√Ωch teplota p≈ôes√°hla 60: ")
vysok√°_teplota_evropske_staty.head()
```

5. Dotaz na extr√©mn√≠ hodnoty, tj. mƒõ≈ôen√≠, ve kter√Ωch je teplota vy≈°≈°√≠ ne≈æ 80 stup≈à≈Ø nebo men≈°√≠ ne≈æ -20 stup≈à≈Ø.

```python
extremni_teploty = seznam[
    (seznam["AvgTemperature"] > 60) | (seznam["AvgTemperature"] < -20)
]
print("\n St√°ty u kter√Ωch byla namƒõ≈ôen√° extr√©mn√≠ teplota: ")
extremni_teploty.iloc[510:515]
```

Ve sv√©m programu nejprve proveƒè import modulu pytemperature. Nov√Ω sloupec pak p≈ôid√°≈° do tabulky tak, ≈æe nalevo od = vlo≈æ√≠≈° tabulku a n√°zev nov√©ho sloupce do hranat√Ωch z√°vorek. Napravo pak m≈Ø≈æe≈° prov√°dƒõt v√Ωpoƒçty pomoc√≠ ji≈æ existuj√≠c√≠ch sloupc≈Ø.
```python
import pytemperature
seznam["AvgTemperatureCelsia"] = pytemperature.f2c(seznam["AvgTemperature"])
seznam.head(1)
```

2. Dotaz na mƒõ≈ôen√≠, ve kter√Ωch je teplota vy≈°≈°√≠ ne≈æ 15 stup≈à≈Ø Celsia a souƒçasnƒõ bylo mƒõ≈ôen√≠ provedeno v regionu (sloupec Region) Evropa (Europe).

3. Dotaz na extr√©mn√≠ hodnoty, tj. mƒõ≈ôen√≠, ve kter√Ωch je teplota vy≈°≈°√≠ ne≈æ 30 stup≈à≈Ø Celsia nebo men≈°√≠ ne≈æ -10 stup≈à≈Ø. Jsou nƒõkter√© hodnoty podez≈ôel√©?
