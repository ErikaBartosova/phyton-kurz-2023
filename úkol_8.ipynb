1. Vypiš si prvních několik řádků, ať si prohlédneš strukturu tabulky.

```python
import pandas as pd
seznam = pd.read_csv("temperature.csv", sep=",")
seznam.head()
```

2. Dotaz na měření, která byla provedena v Praze. Je na datech něco zvláštního?

```python
praha = seznam[seznam["City"] == "Prague"]
print("\n Hodnoty naměřené v Praze jsou: ")
praha.head()
```

3. Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

```python
teplota_nad_80 = seznam[seznam["AvgTemperature"] > 80]
print("\n Měření u kterých byla teplota vyšší něž 80: ")
teplota_nad_80.head()
```

4. Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
vysoká_teplota_evropske_staty = seznam[
    (seznam["AvgTemperature"] > 60) & (seznam["Region"] == "Europe")
]
print("\n Evropské státy u kterých teplota přesáhla 60: ")
vysoká_teplota_evropske_staty.head()
```

5. Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů.

```python
extremni_teploty = seznam[
    (seznam["AvgTemperature"] > 60) | (seznam["AvgTemperature"] < -20)
]
print("\n Státy u kterých byla naměřená extrémní teplota: ")
extremni_teploty.iloc[510:515]
```

Ve svém programu nejprve proveď import modulu pytemperature. Nový sloupec pak přidáš do tabulky tak, že nalevo od = vložíš tabulku a název nového sloupce do hranatých závorek. Napravo pak můžeš provádět výpočty pomocí již existujících sloupců. Můžeš např. použít funkci f2c z modulu pytemperature, která převede teplotu ze stupňů Fahrenheita na stupně Celsia.

```python
import pytemperature
seznam["AvgTemperatureCelsia"] = pytemperature.f2c(seznam["AvgTemperature"])
seznam.head(1)
```

1. Dotaz na měření, ve kterých je teplota (sloupec AvgTemperatureCelsia) vyšší než 30 stupňů Celsia.

```python
stupně_nad_30 = seznam[seznam["AvgTemperatureCelsia"] > 30]
print("\n Měření u kterých teplota přesáhla 30 stupňů Celsia: ")
stupně_nad_30.head()
```

2. Dotaz na měření, ve kterých je teplota vyšší než 15 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
stupně_nad_15_eu = seznam[
    (seznam["AvgTemperatureCelsia"] > 15) & (seznam["Region"] == "Europe")
]
print("\n Evropské státy u kterých teplota přesáhla 15 stupňů Celsia: ")
stupně_nad_15_eu.head()
```

3. Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 30 stupňů Celsia nebo menší než -10 stupňů. Jsou některé hodnoty podezřelé?

```python
extremni_teploty_celsia = seznam[
    (seznam["AvgTemperatureCelsia"] > 30) | (seznam["AvgTemperatureCelsia"] < -10)
]
print("\n Hodnoty u kterých bylo naměřeno více než 30 stupňů Celsia nebo méně než -10 stupňů: ")
extremni_teploty_celsia.head()
```
